# Visualizeing Data With the Help of Graphs using Library's such as Seaborn 

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(df_copy['rating'])
sns.countplot(df_copy['rating']).set_xticklabels(sns.countplot(df_copy['rating']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Rating');
```

# Ratings of Movies and Shows

```python
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=df_copy)
plt.title('Relation between Type and Rating of Movies and Shows Seperately')
plt.show();
```

# Relation between Type and Rating of Movies and Shows

```python
sns.countplot(df_copy['type'])
fig.set_size_inches(10,10)
plt.title('Type');
```

# Relation between Movies and TV Shows

```python
most_titles_produced = df_copy.groupby('country').count().sort_values('title',ascending=False).head(5)

most_titles_produced.reset_index(inplace=True)
most_titles_produced
```

# country has the most number of titles produced

```python
plt.bar(most_titles_produced.country,most_titles_produced.title);
```

# Netflix have more Movies or TV Shows

```python
sns.countplot('type', data=df_copy);
```

# What are the top 5 most popular ratings on Netflix

```python
df_copy_rating = df_copy['rating'].value_counts()
df_copy_rating = pd.DataFrame(df_copy_rating).reset_index()
df_copy_rating.columns = ['rating','Nbr']
sns.barplot(x = 'rating',y = 'Nbr', data=df_copy_rating.head(5));
```

# Which are top 5 least popular genre on Netflix

```python
genre=df_copy["listed_in"].value_counts().tail(5)
genre
```

```python
plt.figure(figsize=(10,4))
genre.plot(kind="bar",color="blue")
plt.title("5 Least Popular Genres on Netflix",size=20);
```

# Which were the top 5 years in number of titles released

```python
top_5_year = df_copy.groupby('release_year').count().sort_values('title',ascending=False).head(5)

top_5_year.reset_index(inplace=True)
top_5_year
```

```python
plt.bar(top_5_year.release_year,top_5_year.title);
```

# Pie-chart for the Type: Movie and TV Shows

```python
labels = ['Movie', 'TV show']
size = df['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

# Pie-chart for Rating

```python
df['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,8))
plt.show()
```

# Most Word Used in Title

```python
from wordcloud import WordCloud
plt.subplots(figsize=(25,15))
wordcloud = WordCloud(
                          background_color='white',
                          width=1920,
                          height=1080
                         ).generate(" ".join(df.title))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('country.png')
plt.show()
```
